### **Outliers Values**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Akan tetapi memang tidak semua kolom perlu kita tangani, kita harus tetap melihat dan memahami tujuan dari setiap kolom yang ada di dataset.

Dan terlihat kolom
* Quantity

Dengan presentase 5%
* Profit

Dengan presentase 3%

* Sales

Dengan presentase 0.6% (dibawah 1%)

````python
results = []

# Check if df is empty before attempting to select dtypes or calculate anything
if df.empty:
    print("DataFrame kosong, tidak dapat menghitung outliers.")
else:
    cols = df.select_dtypes(include=['float64', 'int64'])

    if not cols.empty:
        for col in cols.columns:
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5*iqr
            upper_bound = q3 + 1.5*iqr
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            # Ensure len(df) is not zero before division to avoid ZeroDivisionError
            percent_outliers = (len(outliers)/len(df))*100 if len(df) > 0 else 0
            results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

        # Dataframe dari list hasil
        results_df = pd.DataFrame(results)
        if not results_df.empty:
            results_df.set_index('Kolom', inplace=True)
            results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)
            display(results_df)
        else:
            print("Tidak ada kolom numerik yang valid untuk dianalisis outliers.")
    else:
        print("Tidak ada kolom numerik di DataFrame untuk analisis outliers.")
````

### Analisis Ulang Outliers untuk Kolom Numerik

Sekarang setelah kolom 'Amount' berhasil dikonversi ke tipe data numerik (`float64`), kita akan menjalankan ulang analisis *outliers* untuk memastikan kedua kolom numerik (`Amount` dan `Boxes Shipped`) tercakup.

````python
results = []

if df.empty:
    print("DataFrame kosong, tidak dapat menghitung outliers.")
else:
    # Memilih kolom dengan tipe data float64 dan int64
    cols = df.select_dtypes(include=['float64', 'int64']).columns

    if not cols.empty:
        for col in cols:
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5*iqr
            upper_bound = q3 + 1.5*iqr
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            percent_outliers = (len(outliers)/len(df))*100
            results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

        # Dataframe dari list hasil
        results_df = pd.DataFrame(results)
        if not results_df.empty:
            results_df.set_index('Kolom', inplace=True)
            results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

            # Tampilkan dataframe
            display(results_df)
        else:
            print("Tidak ada kolom numerik yang valid untuk dianalisis outliers setelah pembersihan awal.")
    else:
        print("Tidak ada kolom numerik di DataFrame untuk analisis outliers setelah pembersihan awal.")
````

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas: Membandingkan total nominal penjualan (Sales) di setiap wilayah seluruh dunia (Global).

Tujuan: Mengidentifikasi wilayah mana yang memberikan kontribusi pendapatan terbesar dan membandingkan performa antar wilayah secara langsung.

Visualisasi: Bar Chart (Grafik Batang).::

````python
if not df.empty and 'Country' in df.columns and 'Amount' in df.columns:
    penjualan_country = df.groupby('Country')['Amount'].sum().sort_values(ascending=False)

    plt.figure(figsize=(10,6))
    sns.barplot(x=penjualan_country.index, y=penjualan_country.values, palette='viridis', hue=penjualan_country.index, legend=False)
    plt.title('Total Penjualan Berdasarkan Negara')
    plt.xlabel('Negara')
    plt.ylabel('Total Penjualan')
    plt.xticks(rotation=45)
    plt.show()
elif df.empty:
    print("DataFrame kosong, tidak dapat menampilkan total penjualan berdasarkan negara.")
else:
    print("Kolom 'Country' atau 'Amount' tidak ditemukan atau kosong.")
````

## **Composition/Komposisi**

Aktivitas: Melihat proporsi atau persentase total penjualan berdasarkan kategori produk (Category).

Tujuan: Mengetahui kategori produk mana yang paling mendominasi portofolio penjualan perusahaan secara keseluruhan.

Visualisasi: Pie Chart (Grafik Lingkaran).

````python
if not df.empty and 'Product' in df.columns and 'Amount' in df.columns:
    # Menghitung proporsi produk berdasarkan total penjualan
    sales_by_product = df.groupby('Product')['Amount'].sum().sort_values(ascending=False)
    sales_by_product_top3 = sales_by_product.head(3)

    if not sales_by_product_top3.empty:
        plt.figure(figsize=(10, 8))
        sales_by_product_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
        plt.title('Proporsi 3 Produk Teratas Berdasarkan Total Penjualan')
        plt.ylabel('')
        plt.axis('equal')
        plt.show()
    else:
        print("Tidak ada data produk yang cukup untuk menampilkan pie chart.")
elif df.empty:
    print("DataFrame kosong, tidak dapat menampilkan proporsi produk.")
else:
    print("Kolom 'Product' atau 'Amount' tidak ditemukan atau kosong.")
````

## **Distribution/Distribusi**

## **Relationship/Hubungan**

Aktivitas: Menganalisis korelasi antara nilai penjualan (Sales) dengan keuntungan (Profit).

Tujuan: Mengetahui apakah peningkatan penjualan selalu berbanding lurus dengan peningkatan profit, serta mengidentifikasi transaksi yang memiliki penjualan tinggi namun justru menghasilkan kerugian.

Visualisasi: Heatmap

````python
if not df.empty and 'Amount' in df.columns and 'Boxes Shipped' in df.columns:
    # Korelasi hanya bisa dihitung untuk kolom numerik
    numeric_cols = df[['Amount', 'Boxes Shipped']].select_dtypes(include=['float64', 'int64'])
    if not numeric_cols.empty and numeric_cols.shape[1] > 1:
        plt.figure(figsize=(8, 6))
        sns.heatmap(data=numeric_cols.corr(),
                    annot=True,
                    cmap='viridis',
                    fmt='.2f')
        plt.title('Korelasi antara Amount dan Boxes Shipped')
        plt.show()
    else:
        print("Tidak cukup kolom numerik yang valid untuk menghitung korelasi.")
elif df.empty:
    print("DataFrame kosong, tidak dapat menampilkan heatmap korelasi.")
else:
    print("Kolom 'Amount' atau 'Boxes Shipped' tidak ditemukan atau kosong.")
````

### **Outliers**

````python
# Menangani outlier pada kolom numerik yang valid: 'Amount' dan 'Boxes Shipped'
columns_to_impute = ['Amount', 'Boxes Shipped']

if df.empty:
    print("DataFrame kosong, tidak dapat menangani outlier.")
else:
    for col in columns_to_impute:
        if col in df.columns and df[col].dtype in ['float64', 'int64']:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Melakukan clipping (pembatasan) nilai agar berada dalam rentang lower dan upper bound
            df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
            print(f'Outlier pada kolom {col} telah ditangani dengan clipping.')
        else:
            print(f'Kolom {col} tidak ditemukan atau bukan tipe numerik, tidak dapat menangani outlier.')
````

````python
columns_to_impute = ["Amount", "Boxes Shipped"]

if df.empty:
    print("DataFrame kosong, tidak dapat menangani outlier dengan clipping.")
else:
    for col in columns_to_impute:
        if col in df.columns and df[col].dtype in ['float64', 'int64']:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)

            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
            print(f"Penanganan outlier pada kolom {col} berhasil dilakukan.")
        else:
            print(f"Kolom {col} tidak ditemukan atau bukan tipe numerik, tidak dapat menangani outlier.")
````

````python
results = []

if df.empty:
    print("DataFrame kosong, tidak dapat menghitung persentase outliers.")
else:
    cols = df.select_dtypes(include=['float64', 'int64'])

    if not cols.empty:
        for col in cols.columns:
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5*iqr
            upper_bound = q3 + 1.5*iqr
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            percent_outliers = (len(outliers)/len(df))*100
            results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

        # Dataframe dari list hasil
        results_df = pd.DataFrame(results)
        if not results_df.empty:
            results_df.set_index('Kolom', inplace=True)
            results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

            # Tampilkan dataframe
            display(results_df)
        else:
            print("Tidak ada kolom numerik yang valid untuk dianalisis outliers.")
    else:
        print("Tidak ada kolom numerik di DataFrame untuk analisis outliers.")
````

Dilakukan penyesuaian nilai ekstrem pada kolom Amount dan Boxes Shipped agar hasil analisis lebih akurat:

Metode IQR: Menentukan batas wajar data menggunakan teknik Interquartile Range.

Teknik Clipping: Mengubah nilai yang terlalu tinggi atau rendah ke batas terdekat agar tidak merusak rata-rata.

Data Tetap Utuh: Tidak ada data yang dihapus, sehingga jumlah baris tetap konsisten namun lebih stabil secara statistik.
